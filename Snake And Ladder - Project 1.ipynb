```python
import random as rd

def roll_dice():
    return rd.choice([1, 2, 3, 4, 5, 6])

def check(position):
    ladders = {4: 31, 8: 38, 21: 48, 36: 54, 48: 84, 51: 67, 71: 82, 75: 86, 80: 99}
    snakes = {11: 4, 22: 7, 33: 9, 54: 40, 61: 17, 73: 54, 85: 47, 89: 69, 98: 81}

    if position in ladders:
        print("Yay! You climbed a ladder to", ladders[position])
        position = ladders[position]
    elif position in snakes:
        print("Oops! Snake bites! You go down to", snakes[position])
        position = snakes[position]

    return position

def playing_game():
    # Initial positions for both players
    player1_position = 0
    player2_position = 0
   # Variable to keep track of the current player
    player = 1
   # Loop until either player wins the game
    while True:
        if player == 1:
            input("Player 1, press Enter to roll the dice:")
            rolling_dice = roll_dice()
            print("Player 1 rolled a", rolling_dice)

            if rolling_dice == 6:
                print("Player 1 gets an extra chance!")

            previous_position = player1_position
            player1_position += rolling_dice
            player1_position = check(player1_position)

            if player1_position > 100:
                print("Oops! You exceeded 100. Going back to the previous position.")
                player1_position = previous_position

            print("Player 1 is now at position", player1_position)

            if player1_position == 100:
                print("Player 1 wins the game!")
                break

            if rolling_dice != 6:
                player = 2
        else:
            input("Player 2, press Enter to roll the dice:")
            rolling_dice = roll_dice()
            print("Player 2 rolled a", rolling_dice)

            if rolling_dice == 6:
                print("Player 2 gets an extra chance!")

            previous_position = player2_position
            player2_position += rolling_dice
            player2_position = check(player2_position)

            if player2_position > 100:
                print("Oops! You exceeded 100. Going back to the previous position.")
                player2_position = previous_position

            print("Player 2 is now at position", player2_position)

            if player2_position == 100:
                print("Player 2 wins the game!")
                break

            if rolling_dice != 6:
                player = 1

# Start the game when a player rolls a 6
while True:
    input("Press Enter to roll the dice and start the game:")
    rolling_dice = roll_dice()
    print("You rolled a", rolling_dice)

    if rolling_dice == 6:
        print("Game starting...")
        playing_game()
        break
    else:
        print("You need to roll a 6 to start the game.")
```
